```python
#q-4
import numpy as np
r=int(input("Enter the rows : "))
c=int(input("Enter the columns : "))
key=int(input("1. For all ones\n2. For all zeroes\n3.Random values within a range\n4.Diagonal matrix\n5.for exit"))
if(key==1):
    arr=np.ones((r,c),dtype=int)
    print(arr)
elif key==2:
    arr=np.zeros((r,c),dtype=int)
    print(arr)
elif key==3:
    rng=int(input("Enter the range : "))
    arr=np.random.randint(rng,size=(r,c))
    print(arr)
elif key==4:
    rng=int(input("Enter the range : "))
    x=np.arange(rng).reshape(r,c)
    print(x)
    arr=np.diag(x)
    print(arr)
else:
    print("Exit...")
```

```python
import numpy as np
import matplotlib.pyplot as plt
df['sin_sepal_length'] = np.sin(df['sepal length (cm)'])
df['cos_sepal_width'] = np.cos(df['sepal width (cm)'])

# Scatter plot based on sine and cosine values
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sn.scatterplot(x='sin_sepal_length', y='sepal width (cm)', hue='target', data=iris)
plt.title('Scatter Plot: Sin Sepal Length vs Sepal Width')

plt.subplot(1, 2, 2)
sn.scatterplot(x='sepal length (cm)', y='cos_sepal_width', hue='target', data=iris)
plt.title('Scatter Plot: Sepal Length vs Cos Sepal Width')

plt.tight_layout()
plt.show()
```
